```python
import requests
import json
```

```python
doi = []
IDX=25

for  i in range(43):
  URL="https://api.elsevier.com/content/search/scopus?query=AFFILCOUNTRY(thailand)%20AND%20AFFILORG(chulalongkorn%20university)%20AND%20PUBYEAR%20<%202018&apiKey=2833520c7ecf2922cf7243e5c3f2c4fc&start="+str(IDX*i)+"&httpAccept=application/json"
  response = requests.get(URL)

  if response.status_code == 200:
      for j in range(25):
          if 'prism:doi' in response.json()["search-results"]['entry'][j]:
              data = response.json()["search-results"]['entry'][j]['prism:doi']
              doi.append(data)
  else:
      print('Failed to retrieve data:', response.status_code)
```

```python
apiKey = "843841f3d3ce20809c773386c734377b"
for i in range(len(doi)):
    articleData = requests.get('https://api.elsevier.com/content/abstract/DOI:'+ doi[i] +'?apiKey=' + apiKey + '&view=FULL&httpAccept=application/json')
    with open ('data/'+str(i)+'.json', 'w') as f:
        json.dump(articleData.json(), f)
```
